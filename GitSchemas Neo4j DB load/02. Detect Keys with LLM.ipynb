```python
from Neo4jDriver import Neo4jDriver

with Neo4jDriver() as neo4j:

    get_schema_text_descriptions =  """
        // Count schema with certain characterisitcs
        MATCH (s:Schema)
        OPTIONAL MATCH (s)-->(:Table)-->(:Column)-[:FK_COLUMN]-(fk:ForeignKey)
        OPTIONAL MATCH (s)-->(:Table)-->(:Column)-[:PK_COLUMN]-(pk:PrimaryKey)
        WITH s, count(distinct fk) as numFKs, count(distinct pk) as numPKs
        WITH s, count{(s)-[:CONTAINS_TABLE]->(:Table)} as numTables, numFKs, numPKs
        WHERE numTables > 10 AND numPKs = numTables AND numFKs > numPKs
        WITH s
        MATCH (s)-[:CONTAINS_TABLE]->(t:Table)-[:HAS_COLUMN]->(c:Column)
        OPTIONAL MATCH (c)<-[:PK_COLUMN]-(pk:PrimaryKey)
        OPTIONAL MATCH (c)<-[:FK_COLUMN]-(fk:ForeignKey)
        WITH s,
            t.name as tableName,
            collect({name: c.name,
                    pk: CASE (not pk is null and $printKeyInfo) WHEN True THEN '(PK)' ELSE '' END, 
                    fk: CASE (not pk is null and $printKeyInfo) WHEN True THEN '(FK)' ELSE '' END
                    }) as columns
        WITH s, tableName, [x in columns | x.name + x.fk + x.pk] as columns
        WITH s, 'Table "' + tableName + '" has columns:\n\n' + apoc.text.join(columns,'\n') as tableDescriptions
        WITH s, apoc.text.join(collect(tableDescriptions),'\n---------------------\n\n') as schemaDescription
        RETURN s.url, schemaDescription
        """

    schemaDescriptions = neo4j.execute_read_query(get_schema_text_descriptions, {'printKeyInfo': False})
    # result = neo4j.execute_read_query("RETURN 1 as a")
    
display(schemaDescriptions)
```
